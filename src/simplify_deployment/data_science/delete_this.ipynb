```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
openstaand = 300000
rente_per_maand = 0.03/12
afbetaling = 800
totaal_betaalde_rente = 0
openstaand_lijst = []
```

```python
maand = 1
while openstaand >= 0:
    openstaand_lijst.append(openstaand)
    te_betalen_rente = rente_per_maand*openstaand
    totaal_betaalde_rente += te_betalen_rente
    openstaand = openstaand - (afbetaling - te_betalen_rente)
    print(f"Openstaand bedrag is {openstaand} euro.")
    print(f"Totale rente betaald: {totaal_betaalde_rente}")
    print(f"Maand: {maand}. Jaar: {maand/12}")
    maand += 1
```

```python
plt.plot(openstaand_lijst)
```

```python
x = np.arange(1111)
y = 300000*(1+rente_per_maand/12)**x-x*afbetaling
plt.plot(x,y)
```
