```python
import pandas as pd
from datetime import timedelta
```

```python
minute = pd.read_parquet("/home/thomas/repos/simplify_deployment/data/minute_data_simplify_study.parquet")
minute
```

```python
stop_test = max(minute.index)
start_test = stop_test - timedelta(weeks = 4)

stop_train = start_test
start_train = stop_train - timedelta(weeks = 16)
```

```python
minute_train = minute.loc[start_train:stop_train]
minute_test = minute.loc[start_test:stop_test]
minute_train.fillna(0).to_parquet("/home/thomas/repos/simplify_deployment/data/minute_train.parquet")
minute_test.fillna(0).to_parquet("/home/thomas/repos/simplify_deployment/data/minute_test.parquet")
```
